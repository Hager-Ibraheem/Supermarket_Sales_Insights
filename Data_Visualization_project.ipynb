# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

# Read Data

```python
df=pd.read_csv("supermarket_sales - Sheet1.csv")
df.head()
```

```python
# Data information
df.info()
```

```python
#confirm of cleaning
df.isnull().sum()
```

## 1. What are the most sold items?

```python
prod_sold=df.groupby('Product line')['Quantity'].sum().reset_index()
# Grouped data and using sum method to have quantity of each product.
prod_sold
```

```python
prod_sold_fig=px.pie(prod_sold,names='Product line',values='Quantity',title='Quantity of each product sold'
                   ,color_discrete_sequence=px.colors.sequential.RdBu ,hole=0.2)
#apply specific colors using 'color_discrete_map' to have a sequence of colors and make the hole of pie using hole method.
prod_sold_fig.show()
```

## 2. Which products are more sold by Male and female?

```python
#group data by two colums based on sum of Quantity of items sold .
item_gender=df.groupby(['Product line','Gender'],as_index=False).Quantity.sum()
item_gender
```

```python
#specify colors for Gender column
Gender_color_map={'Male':'rgb(0,0,0)','Female':'rgb(178,34,34)'}
#apply line chart to determinr num of items sold by both kinds of genders
item_gender_fig=px.line(item_gender,
                        x='Product line',
                        y='Quantity',
                        color_discrete_map=Gender_color_map,
                        color='Gender',#customize colors depends on 'Gender'
                        title='Products purchased by Male vs Female customers'
                    )
item_gender_fig.show()
```

```python
#apply bar chart to determinr num of items sold by both kinds of genders
item_gender_fig1=px.bar(item_gender,
                        x='Gender',
                        y='Quantity',
                       barmode='group',
                        color_discrete_sequence=px.colors.sequential.RdBu, #set colors for graph
                        color='Product line',#customize colors depends on 'product line'
                        title='Products purchased by Male vs Female customers'
                    )
item_gender_fig1.show()
```

## 3. Which branch has lower gross income?

```python
#plot box plot between gross income and branch columns to know which branch has lower gross income.
#use the box plot because it's an easy way to compare the interquartile ranges (that is, the box lengths).
color_scale={'A':'rgb(255,0,127)','B':'rgb(255,153,204)','C':'rgb(51,0,25)'} #dictionary mapping specific categorical values to colors using a string RGB code specification.
FIG1=px.box(df,
           x='Branch',y='gross income', title='Gross Income By Every Branch.',
           color_discrete_map=color_scale, #apply specific colors using 'color_discrete_map' argument.
           color='Branch' #customize colors depends on 'Branch' column which each category in 'Branch' column take specific color.
           )
FIG1.update_xaxes(categoryorder='total descending') #Sort box plots in descending order.
FIG1.show() #display the plot.
```

## 4. How is the overall performance of different branches for products line?

```python
item_data=df.groupby(['City','Product line'])['Total'].sum().reset_index()
# Grouped data to have the sum of each product.
item_data
```

```python
item_bar=px.bar(item_data,x='Product line',y='Total',color='City',barmode='group',title='Sales generated by every item in different cities',color_discrete_sequence=px.colors.sequential.Plasma)
#apply specific colors using 'color_discrete_map' to have a sequence of colors.
item_bar.update_xaxes(categoryorder='total descending') #Sort bars in descending order.
item_bar.show()
```

## 5.How is the performance of supermarket during the three months ?

```python
#use Pandas to_datetime() method to convert string Date time into Python Date time object so we can perform operations on Date Time object rather a string object.
df['Date']=pd.to_datetime(df['Date'])
df.head()
```

```python
#group data based on sum of gross income (calculate sum of gross income for each day).
sum_date_income=df.groupby('Date')['gross income'].sum().reset_index()
sum_date_income
```

```python
#plot line chart for Date and gross income (gross income during 3 months).
#use the line chart because it's a good way to show data trends.
FIG2=px.line(sum_date_income,
             x='Date',y='gross income', title='Gross Income During Three Months',
             color_discrete_sequence=px.colors.sequential.Viridis #customize colors using Built in Sequential Color scales.
             )
FIG2.show() #display the plot.
```

## 6. How is the overall performance of different branches each day during the three months?

```python
#calculate sales in every branch for each day.
sum_branch_sales=df.groupby(['Date','Branch'])['Total'].sum().reset_index()
sum_branch_sales
```

```python
#plot line chart for sales in every branch during 3 months.
#use the line chart because it's a good way to show data trends.
FIG3=px.line(sum_branch_sales,
             x='Date',y='Total', title='Sales In Every Branch During Three Months',
             color_discrete_sequence=px.colors.sequential.Blackbody, #customize colors using Built in Sequential Color scales.
             color='Branch' #customize colors depends on 'Branch' column which each category in 'Branch' column take specific color.
             )
FIG3.show() #display the plot.
```

## Sales in every branch per month

```python
#plot histogram for sales in every branch per month.
#use histogram because in plotly it's a way to show different values in one column by using the color argument(Stacked values).
#use histogram because in plotly it's a good way to visualize data on categorical and date axes.
FIG4=px.histogram(sum_branch_sales,
             x='Date',y='Total', title='Sales In Every Branch Per Month',
                  nbins=3, #choose the number of bins.
              color_discrete_sequence=px.colors.sequential.RdBu, #customize colors using Built in Sequential Color scales.
             color='Branch' #customize colors depends on 'Branch' column which each category in 'Branch' column take specific color.
                 )
FIG4.update_layout(bargap=0.1) #make gap between bars(bins).
FIG4.show() #display the plot.
```

```python
#plot histogram for sales in every branch per month.
#use histogram because in plotly it's a good way to visualize data on categorical and date axes.
FIG4=px.histogram(sum_branch_sales,
             x='Date',y='Total', title='Sales In Every Branch Per Month',
                  nbins=3, #choose the number of bins.
              color_discrete_sequence=px.colors.sequential.RdBu, #customize colors using Built in Sequential Color scales.
             color='Branch', #customize colors depends on 'Branch' column which each category in 'Branch' column take specific color.
              barmode='group' #change the default stacked bars to grouped bars.
                 )
FIG4.update_layout(bargap=0.1) #make gap between bars(bins).
FIG4.show() #display the plot.
```
